 Total number of births for each unique day of the week.

```python
from collections import defaultdict
week_birth = defaultdict( list )
for row in input_file:
    week_birth[ row['day_of_week']].append(int(row['births']))
```

```python
cdc_day_births = { k:sum(v) for k,v in week_birth.items()}
cdc_day_births
```

functions to calculate the total birth per N ( month/week/year )

```python
def calc_counts( input_file, column ):
    d = defaultdict( list )
    for row in input_file:
        d[ row[column]].append(int(row['births']))
    return { k:sum(v) for k,v in d.items() }
```

```python
def run():
    input_file = list( csv.DictReader(open('US_birth.csv')))
    cdc_day_births=calc_counts( input_file, 'day_of_week')
    cdc_year_births = calc_counts( input_file, 'year')
    print(cdc_year_births)
```

```python
run()
```
